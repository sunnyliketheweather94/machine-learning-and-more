## Fixed-Point of `cos(x)`

```python
import os
import sys

sys.path.append(os.path.abspath(os.path.join(os.path.abspath("."), os.pardir)))

import numpy as np
from rl.iterate import iterate, converge
```

```python
x = 1.0

values = converge(
    values=iterate(step=lambda x: np.cos(x), start=x),
    done=lambda a, b: np.abs(a - b) < 1e-3,
)

for i, v in enumerate(values):
    print(f"{i + 1:2d}: {v:.4f}")
```
