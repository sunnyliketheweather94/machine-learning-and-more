## Transformations of Random Variables

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
```

```python
f1 = lambda x: x**2
f2 = lambda x: np.sin(2 * np.pi * x)
f3 = lambda x: np.exp(x)
```

```python
xx = ss.uniform(0, 1).rvs(size=10_000)

sns.histplot(x=xx, kde=True)
plt.show()
```

```python
_, axes = plt.subplots(3, 1, figsize=(6, 10))

yy1 = f1(xx)
yy2 = f2(xx)
yy3 = f3(xx)

sns.kdeplot(x=yy1, label="y = x ** 2", ax=axes[0])
sns.kdeplot(x=yy2, label="y = sin(2 * pi * x)", ax=axes[1])
sns.kdeplot(x=yy3, label="y = exp(x)", ax=axes[2])

plt.legend()
plt.tight_layout()
plt.show()
```
